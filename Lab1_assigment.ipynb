```python
from qiskit.visualization import plot_bloch_vector
from math import pi
```

Question 1

$\frac{1}{\sqrt2}(\vert0\rangle+\vert1\rangle)$

for this $\theta$ = $\frac{\pi}{2}$ and $\phi$ = 0

```python
plot_bloch_vector([1,pi/2,0],coord_type='spherical')
```

Question 2

$\frac{1}{\sqrt2}(\vert0\rangle+i\vert1\rangle)$

for this $\theta$ = $\frac{\pi}{2}$ and $\phi$ = $\frac{\pi}{2}$

```python
plot_bloch_vector([1,pi/2,pi/2],coord_type='spherical')
```

Question 3

  $\frac{1}{\sqrt2}\begin{bmatrix}i\\1\end{bmatrix}$
  
= $\frac{1}{\sqrt2}i\vert0\rangle + \frac{1}{\sqrt2}\vert1\rangle$

=$i(\frac{1}{\sqrt2}\vert0\rangle - \frac{1}{\sqrt2}i\vert1\rangle)$

thus global phase = $\frac{\pi}{2}$ , $\theta$ = $\frac{\pi}{2}$ and $\phi$ = $3\frac{\pi}{2}$

```python
plot_bloch_vector([1,pi/2,3*pi/2],coord_type='spherical')
```

Question 4

  $\frac{1}{\sqrt2}\begin{bmatrix}-1\\1\end{bmatrix}$

= $-1(\frac{1}{\sqrt2}\vert0\rangle - \frac{1}{\sqrt2}\vert1\rangle)$

Thus Global Phase = $\pi$ , $\theta$ = $\frac{\pi}{2}$ and $\phi$ = $\pi$

```python
plot_bloch_vector([1,pi/2,pi],coord_type='spherical')
```

Question 5

  $\frac{1}{2}\begin{bmatrix}1+i\\1-i\end{bmatrix}$

= $\frac{1+i}{\sqrt2}(\frac{1}{\sqrt2}\vert0\rangle - \frac{1}{\sqrt2}i\vert1\rangle)$

Thus Global Phase = $\pi$/4 , $\theta$ = $\pi$/2 and $\phi$ = 3*$\pi$/2

```python
plot_bloch_vector([1,pi/2,3*pi/2],coord_type='spherical')
```
